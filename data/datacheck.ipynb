# Collision Prevention: Calculation
1. input: SHIP_ID, RECPTN_DT(datetime), time length (optional)
2. calculate circular area of focus: according to SHIP_ID, RECPTN_DT
  - center: Latitude and Longitude point of the SHIP_ID at RECPTN_DT
  - radius: 1/6 SOG value of the vessel of the SHIP_ID at RECPTN_DT (default)
  - 1 knot = 1nm / 1hour  (nm: nautical mile, 1nm = 1' latitude = 1.852 km)
    - if other specific time length is mentioned in input, calculate new radius
    - e.g. 40 min: radius = SOG value * (40/60)
3. split the circular area into 6 parts: according to COG
  - coordinates axis: COG
  - area1: +30 degrees & -30 degrees from the coordinate axis (60 degree arc)
  - area2 ~ 6: 60 degree arc each, with starting angle +60 of area1
4. calculate the number of all the other ships inside each area: according to RECPTN_DT
  - timeline to focus: 10 min from input RECPTN_DT (default)
    - if other specific time length is mentioned in input, calculate new timeline
    - e.g. 40 min: 2022-08-24 19:09:00 ~ 2022-08-24 19:49:00
  - add up the numbers of unique ships that pass through each area at least once
    - e.g. ship with id '308628000' is detected in area 2 during the timestamp: add 1 count to area2
    - if a same ship passes multiple areas during the expectied timeline, add 1 count for each area
    - e.g. ship with id '308628000' is detected in area 2,3,4 during the timestamp: add 1 count to area2,3,4 each
5. return the unique ship counts and ship ids for each area
  - area color: red (max count), green (none), yellow 
  - e.g. {'area1': {'count': 0, 'ship_ids': []}, 'area2': {'count': 0, 'ship_ids': []}, 'area3': {'count': 0, 'ship_ids': []}, 'area4': {'count': 4, 'ship_ids': ['440059890', '440105630', '440109400', '441198000']}, 'area5': {'count': 1, 'ship_ids': ['440502500']}, 'area6': {'count': 0, 'ship_ids': []}}

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
import pandas as pd
from datetime import timedelta
import math
import numpy as np


# Function to read the GeoJSON file
def checkGeoJson(filename):
    # Path to the GeoJSON file
    geojson_data_path = '/data1/2024_KRISO/yykim/data_ais/' + filename + '.geojson'
    
    # Read the GeoJSON file
    try:
        gdf_geojson = gpd.read_file(geojson_data_path)
        print("GeoJSON data loaded successfully.")
        display(gdf_geojson.head())  # Display the first few rows of the GeoDataFrame
        return gdf_geojson
    except Exception as e:
        print(f"An error occurred while loading the GeoJSON data: {e}")


# Function to calculate ships around an input ship
def calculate_ships_around(df, ship_id, recptn_dt, time_length=60):
    # Ensure RECPTN_DT is datetime
    df['RECPTN_DT'] = pd.to_datetime(df['RECPTN_DT'])
    
    # Get the specific row of the input ship at the given timestamp
    ship_row = df[(df['SHIP_ID'] == ship_id) & (df['RECPTN_DT'] == recptn_dt)].iloc[0]
    ship_position = Point(ship_row['LO'], ship_row['LA'])
    sog = ship_row['SOG']
    cog = ship_row['COG']
    
    # Calculate radius based on SOG and time length
    radius = sog * (time_length / 60)

    print('center(LO,LA): ', ship_position)
    print('radius: ', radius)
    
    # Define the time window
    start_time = recptn_dt - timedelta(minutes=time_length / 2)
    end_time = recptn_dt + timedelta(minutes=time_length / 2)

    print('timeline: ', start_time, ' ~ ' , end_time)
    
    # Filter ships within the time window
    filtered_df = df[(df['RECPTN_DT'] >= start_time) & (df['RECPTN_DT'] <= end_time)]
    
    # Function to create circular sector
    def create_sector(center, radius, start_angle, end_angle, num_points=100):
        angles = np.linspace(start_angle, end_angle, num_points)
        points = [
            Point(
                center.x + radius * math.cos(math.radians(angle)),
                center.y + radius * math.sin(math.radians(angle))
            )
            for angle in angles
        ]
        return Polygon([center] + points + [center])
    
    # Create 6 sectors
    sectors = []
    for i in range(6):
        start_angle = cog - 30 + i * 60
        end_angle = start_angle + 60
        sectors.append(create_sector(ship_position, radius, start_angle, end_angle))
    
    # Count ships in each sector
    result = {}
    for i, sector in enumerate(sectors):
        area_ships = filtered_df[filtered_df['geometry'].apply(lambda x: sector.contains(x))]
        unique_ships = area_ships['SHIP_ID'].unique()
        result[f'area{i+1}'] = {
            'count': len(unique_ships),
            'ship_ids': unique_ships.tolist()
        }
    
    return result
```

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString
import pandas as pd
from datetime import timedelta
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import RegularPolygon

# Function to read the GeoJSON file
def checkGeoJson(filename):
    # Path to the GeoJSON file
    geojson_data_path = '/data1/2024_KRISO/yykim/data_ais/' + filename + '.geojson'
    
    # Read the GeoJSON file
    try:
        gdf_geojson = gpd.read_file(geojson_data_path)
        print("GeoJSON data loaded successfully.")
        display(gdf_geojson.head())  # Display the first few rows of the GeoDataFrame
        return gdf_geojson
    except Exception as e:
        print(f"An error occurred while loading the GeoJSON data: {e}")

# Function to create circular sector
def create_sector(center, radius, start_angle, end_angle, num_points=100):
    angles = np.linspace(start_angle, end_angle, num_points)
    points = [
        Point(
            center.x + radius * math.cos(math.radians(angle)),
            center.y + radius * math.sin(math.radians(angle))
        )
        for angle in angles
    ]
    return Polygon([center] + points + [center])

# Function to calculate ships around an input ship
def calculate_ships_around(df, ship_id, recptn_dt, time_length=10):
    # Ensure RECPTN_DT is datetime
    df['RECPTN_DT'] = pd.to_datetime(df['RECPTN_DT'])

    print('ship id: ', ship_id)
    
    # Get the specific row of the input ship at the given timestamp
    ship_row = df[(df['SHIP_ID'] == ship_id) & (df['RECPTN_DT'] == recptn_dt)].iloc[0]
    ship_position = Point(ship_row['LO'], ship_row['LA'])
    sog = ship_row['SOG']
    cog = ship_row['COG']
    
    # Cap SOG at 102 knots if it is over 102
    if sog > 102:
        sog = 102
    
    # Calculate radius based on SOG and time length
    radius = sog * (time_length / 60)

    print('center (LO,LA): ', ship_position)
    print('radius (knot): ', radius)
    
    # Define the time window
    start_time = recptn_dt
    end_time = recptn_dt + timedelta(minutes=time_length)

    print('timeline: ', start_time, ' ~ ' , end_time)
    
    # Filter ships within the time window
    filtered_df = df[(df['RECPTN_DT'] >= start_time) & (df['RECPTN_DT'] <= end_time)]
    
    # Create 6 sectors
    sectors = []
    for i in range(6):
        start_angle = cog - 30 + i * 60
        end_angle = start_angle + 60
        sectors.append(create_sector(ship_position, radius, start_angle, end_angle))
    
    # Count ships in each sector
    result = {}
    max_ships = 0
    for i, sector in enumerate(sectors):
        area_ships = filtered_df[filtered_df['geometry'].apply(lambda x: sector.contains(x))]
        unique_ships = area_ships['SHIP_ID'].unique().tolist()
        if ship_id in unique_ships:
            unique_ships.remove(ship_id)  # Remove the input ship from the list
        count_ships = len(unique_ships)
        result[f'area{i+1}'] = {
            'count': count_ships,
            'ship_ids': unique_ships
        }
        if count_ships > max_ships:
            max_ships = count_ships
    
    # Plot the map
    fig, ax = plt.subplots(figsize=(10, 10))
    base = df.plot(ax=ax, color='gray', markersize=5)
    
    # Color sectors based on ship counts
    for i, sector in enumerate(sectors):
        area_info = result[f'area{i+1}']
        if area_info['count'] == 0:
            color = 'green'
        elif area_info['count'] == max_ships:
            color = 'red'
        else:
            color = 'yellow'
        gpd.GeoSeries([sector]).plot(ax=base, color=color, alpha=0.5, edgecolor='black')
    
    # Plot the input ship's position with a triangle marker
    gpd.GeoSeries([ship_position]).plot(ax=base, color='red', markersize=50)
    triangle = RegularPolygon((ship_position.x, ship_position.y), numVertices=3, radius=0.01, orientation=np.radians(cog), color='red')
    ax.add_patch(triangle)
    
    circle = plt.Circle((ship_position.x, ship_position.y), radius, color='blue', fill=False)
    ax.add_artist(circle)
    
    # Add trace lines for detected ships
    for unique_ship_id in filtered_df['SHIP_ID'].unique():
        ship_data = filtered_df[filtered_df['SHIP_ID'] == unique_ship_id]
        if len(ship_data) > 1:  # Ensure there are more than one points to create LineString
            line = LineString([(row['LO'], row['LA']) for idx, row in ship_data.iterrows()])
            if unique_ship_id == ship_id:
                gpd.GeoSeries([line]).plot(ax=base, color='blue', linewidth=2)
            else:
                gpd.GeoSeries([line]).plot(ax=base, color='black', linewidth=1)
    
    plt.title(f'Ship Collision Detection for SHIP_ID {ship_id} at {recptn_dt}')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.grid()
    plt.show()
    
    return result
```
